```python
# define the GRIN coefficients for spherical lens
n0 = 1.37
n1 = -0.01
n3 = 0.04
n4 = -0.01

# define variable z and height func h
z = var("z")
h = function("h")(z)

#define function for spherical GRIN distribution
n(z,h) = n0 + n1*(h^2) + n3*z + n4*z^2 == 1.37

#plot of spherical region
contour_plot(n(z,h) ,(z,-6,10),(h,-6,6), fill = false, contours = 20)
```

```python
# define the GRIN coefficients for elliptic lens
n0 = 1.37
n1 = -0.01
n3 = 0.04
n4 = -0.02

# define variable z and height func h
z = var("z")
h = function("h")(z)

#define function for  elliptic GRIN distribution
n(z,h) = n0 + n1*(h^2) + n3*z + n4*z^2 == 1.37

#plot of anterior region
contour_plot(n(z,h) ,(z,-6,10),(h,-6,6), fill = false, contours = 20)
```
